```python
kc.drop('waterfront',axis=1,inplace=True)
```

```python
kc1=kc[(kc['sqft_lot']>(np.mean(kc['sqft_lot'])-3*np.std(kc['sqft_lot']))) & (kc['sqft_lot']<(np.mean(kc['sqft_lot'])+0.4*np.std(kc['sqft_lot'])))]
kc1.shape
```

```python
kc1.skew()
```

```python
kc2=kc1[(kc1['sqft_lot15']>(np.mean(kc1['sqft_lot15'])-3*np.std(kc1['sqft_lot15']))) & (kc1['sqft_lot15']<(np.mean(kc1['sqft_lot15'])+0.4*np.std(kc1['sqft_lot15'])))]
kc2.shape
```

```python
kc2.skew()
```

```python
kc3=kc2[(kc2['price']>(np.mean(kc2['price'])-3*np.std(kc2['price']))) & (kc2['price']<(np.mean(kc2['price'])+3*np.std(kc2['price'])))]
kc3.skew()
```

```python
a=['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors','view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built','yr_renovated', 'lat', 'long', 'sqft_living15', 'sqft_lot15']
b=1
plt.figure(figsize=([20,20]))
for i in a:
    plt.subplot(6,3,b)
    sns.scatterplot(kc3[i],kc3['price'])
    b+=1
```

```python
from sklearn.model_selection import train_test_split
```

```python
df3_train,df3_test=train_test_split(kc3,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df3_train_sc=pd.DataFrame(sc.fit_transform(df3_train),columns=kc3.columns)
```

```python
df3_test_sc=pd.DataFrame(sc.transform(df3_test),columns=kc3.columns)
```

```python
df3_train_sc.shape
```

```python
df3_test_sc.shape
```

```python
x_train=df3_train_sc.drop('price',axis=1)
```

```python
x_test=df3_test_sc.drop('price',axis=1)
```

```python
y_train=df3_train_sc.price
y_test=df3_test_sc.price
```

```python
from sklearn.feature_selection import RFE
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
ref=RFE(lr,14)
ref=ref.fit(x_train,y_train)
```

```python
list(zip(x_train.columns,ref.support_,ref.ranking_))
```

```python
selected_col=x_train.columns[ref.support_]
selected_col
```

```python
x_train_ref=x_train[selected_col]
```

```python
import statsmodels.api as sm
```

```python
x_train_ref_sm=sm.add_constant(x_train_ref)
```

```python
linmod=sm.OLS(y_train,x_train_ref_sm).fit()
```

```python
linmod.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['Features']=x_train_ref.columns
vif['VIFs']=[variance_inflation_factor(x_train_ref.values, i) for i in range(x_train_ref.shape[1])]

vif['VIFs']=round(vif['VIFs'],2)
vif=vif.sort_values(by='VIFs',ascending=False)
vif
```

```python
x_train_ref_sm.drop('sqft_living',axis=1,inplace=True)
```

```python
linmod1=sm.OLS(y_train,x_train_ref_sm).fit()
linmod1.summary()
```

```python
vif=pd.DataFrame()
vif['Features']=x_train_ref_sm.columns
vif['VIFs']=[variance_inflation_factor(x_train_ref_sm.values, i) for i in range(x_train_ref_sm.shape[1])]

vif['VIFs']=round(vif['VIFs'],2)
vif=vif.sort_values(by='VIFs',ascending=False)
vif
```

```python
y_train_sm_pred=linmod1.predict(x_train_ref_sm)
error=y_train-y_train_sm_pred
sns.distplot(error)
```
